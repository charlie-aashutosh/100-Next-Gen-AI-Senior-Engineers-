```python
import numpy as np
import pandas as pd #data analysis
```

```python
data=pd.read_csv("/content/fraud_transactions_synthetic.csv")
```

```python
data.head() #head() allows us to display the first 5 records of the dataset
```

```python
data.shape #here we have 40,000 rows and 11 columns
#out of 11 columns, 10 columns/features are belonging to the independent category - they are the input features to the NN
#and only 1 is a dependent column
```

```python
import tensorflow as tf
from tensorflow import keras #one of the most important modules within Tensorflow
from tensorflow.keras import layers #layers will help us in developing the neural network layers
```

```python
#divide the data into dependent and independent colums
x=data.drop(['is_fraud'],axis=1) #stores the independent columnns
y=data['is_fraud'] #stores the dependent or the target column
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape
```

```python
x_test.head()
```

```python
x_test.shape
```

```python
x_train.shape[1] #total of 10 neurons are required for us to create the input layer of the NN
```

```python
#building the NN using Tensorflow and Keras - we build it like a Sequential Model with layers stacked on top of each other
model=keras.Sequential([
    layers.Input(shape=(x_train.shape[1],),name='input_layer'), #input_layer has 10 neurons
    layers.Dense(16,activation='relu',name='hidden_layer_1'), #hidden_layer_1 has 16 neurons
    layers.Dense(32,activation='relu',name='hidden_layer_2'), #hidden_layer_2 has 32 neurons
    layers.Dense(1,activation='softmax',name='output_layer') #output_layer has just 1 neuron since it is a binary classifier
])
```

```python
model.compile(
    optimizer='adam', #optimizer will ensure that the gradient loss is within the range
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x_train,y_train,epochs=10,batch_size=32)
```

```python
y_pred=model.predict(x_test)
```
